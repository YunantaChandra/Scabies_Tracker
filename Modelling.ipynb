```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('kucing11_baru.csv')
dataset.tail()
```

```python
dataset['kucing sedang mengalami scabies'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
X = dataset.drop(labels='kucing sedang mengalami scabies', axis=1).values
y = dataset['kucing sedang mengalami scabies'].values
```

```python
np.unique(y, return_counts=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=4)

print ('Train set:', X_train.shape, y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

```python
#Normalisasi
scaler = StandardScaler()
scaler.fit(X_train)

# Standardization Feature
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
print(X_train_scaled[:5, :])
```

```python
#Membangun model SVM
classifier = SVC(kernel = 'linear', random_state = 0)

#Fit model SVM ke data
classifier.fit(X_train, y_train)

#Membuat prediksi
y_pred = classifier.predict(X_test)
```

```python
#Evaluasi Model SVM menggunakan metrices Accuracy
#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

#Evaluasi performa model dengan accuracy
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
model = SVC()
model.fit(X_train_scaled, y_train)
```

```python
y_svm = model.predict(X_test_scaled)
print(classification_report(y_test, y_svm))
```
